# Social network task

## Organize variables

```python
#------------------------------------------------------------------------
# mean preference for each character
#------------------------------------------------------------------------

for role in character_roles:
    
    choice = data[[c for c in data.columns if ('forced_choice' in c) & (f'_{role}_' in c) & ('reaction_time' not in c)]]
    choice = choice.loc[:, [x==role for x in [c.split('_')[-1] for c in choice.columns]]]
    
    data[f'network_choice_{role}_mean'] = np.mean(choice, 1)
    

#------------------------------------------------------------------------
# trial by trial preference
#------------------------------------------------------------------------

choices      = data[[c for c in data.columns if ('forced_choice' in c) & ('reaction_time' not in c)]]
choice_comps = np.unique([('_').join(c.split('_')[2:5]) for c in choices.columns])

for c, comp in enumerate(choice_comps):
    
    
    chars = comp.split('_v_')
    
    if 'neutral' not in chars: # can also try assume 0,0 fr the neutral?

        print(comp)
        
        # chose the 1st character
        mask01  = (data[f'forced_choice_{comp}_{chars[0]}'] > data[f'forced_choice_{comp}_{chars[1]}'])
        pref01  = (data[f'forced_choice_{comp}_{chars[0]}'] * (mask01 * 1))
        affil01 = data[f'affil_mean_{chars[0]}'][mask01]    - data[f'affil_mean_{chars[1]}'][mask01]
        power01 = data[f'power_mean_{chars[0]}'][mask01]    - data[f'power_mean_{chars[1]}'][mask01]
        dist01  = data[f'pov_dist_{chars[0]}'][mask01]      - data[f'pov_dist_{chars[1]}'][mask01]
        cos01  = np.cos(data[f'pov_2d_angle_{chars[0]}'][mask01]) - np.cos(data[f'pov_2d_angle_{chars[1]}'][mask01])
        
        # chose the 2nd character
        mask02  = (data[f'forced_choice_{comp}_{chars[0]}'] < data[f'forced_choice_{comp}_{chars[1]}'])
        pref02  = (data[f'forced_choice_{comp}_{chars[1]}'] * (mask02 * 1))
        affil02 = data[f'affil_mean_{chars[1]}'][mask02]    - data[f'affil_mean_{chars[0]}'][mask02]
        power02 = data[f'power_mean_{chars[1]}'][mask02]    - data[f'power_mean_{chars[0]}'][mask02]
        dist02  = data[f'pov_dist_{chars[1]}'][mask02]      - data[f'pov_dist_{chars[0]}'][mask02]
        cos02   = np.cos(data[f'pov_2d_angle_{chars[1]}'][mask02]) - np.cos(data[f'pov_2d_angle_{chars[0]}'][mask02])
        
        # preferred character, preference amount, affil diff, power diff 
        pref_char = []
        for t in (mask01.values * 1) + (mask02.values * 2):
            if t == 1:   pref_char.append(chars[0])
            elif t == 2: pref_char.append(chars[1])
            else:        pref_char.append(np.nan)

        # TODO: what other info should we have here?
        # TODO: unit test this sort of thing....
        data[f'socialValue_char_pref_{comp}']   = pref_char
        data[f'socialValue_choice_pref_{comp}'] = pref01.values + pref02.values # should all be > 0
        data[f'socialValue_choice_rt_{comp}']   = data[f'forced_choice_{comp}_reaction_time']
        data[f'socialValue_affil_pref_{comp}']  = pd.concat([affil01, affil02]) # 
        data[f'socialValue_power_pref_{comp}']  = pd.concat([power01, power02])
        data[f'socialValue_dist_pref_{comp}']   = pd.concat([dist01, dist02])
        data[f'socialValue_cos_pref_{comp}']    = pd.concat([cos01, cos02])
        
        # if being diff skincolor/gender matters, suggests to followup
        data[f'socialValue_skincolor_{comp}']   = (data[f'{chars[0]}_gender'] == data[f'{chars[1]}_gender']) * 1
        data[f'socialValue_gender_{comp}']      = (data[f'{chars[0]}_skincolor'] == data[f'{chars[1]}_skincolor']) * 1

for var in ['char', 'choice_pref', 'choice_rt', 'affil_pref', 'power_pref', 'dist_pref', 'cos_pref']:
    data[f'socialValue_{var}_avg'] = np.mean(data[[c for c in data.columns if (f'socialValue_{var}_' in c) & ('avg' not in c)]], axis=1)
    
```

## Choice ("valuation"?) reflects affil & power, affil > power
- valuation patterns ~ affil patterns + power patterns
-  characters w/ higher affil should be more valued, across board... 
- but prefernce for powerful relationships might depend on goals etc: some ppl might want subordinate ppl; others might want powerful ppl around them
- basically see this: both greater than 0 in their positive effect on preference, but affil > power

```python
# y = preference patterns
# x = affil patterns + power patterns

ols_df = pd.DataFrame()
for s, sub in data.iterrows():
    
    ols_df.loc[s, 'sub_id'] = sub['sub_id']

    prefs  = sub[[c for c in sub.index if ('socialValue_choice_pref_' in c) & ('avg' not in c)]].values.astype(float)
    rts    = sub[[c for c in sub.index if ('socialValue_choice_rt_' in c) & ('avg' not in c)]].values.astype(float)
    affil  = sub[[c for c in sub.index if ('socialValue_affil_pref_' in c) & ('avg' not in c)]].values.astype(float)
    power  = sub[[c for c in sub.index if ('socialValue_power_pref_' in c) & ('avg' not in c)]].values.astype(float)
    dist   = sub[[c for c in sub.index if ('socialValue_dist_pref_' in c) & ('avg' not in c)]].values.astype(float)
    cosine = sub[[c for c in sub.index if ('socialValue_cos_pref_' in c) & ('avg' not in c)]].values.astype(float)   

    race   = sub[[c for c in sub.index if ('socialValue_skincolor_' in c) & ('avg' not in c)]].values.astype(float)
    gender = sub[[c for c in sub.index if ('socialValue_gender_' in c) & ('avg' not in c)]].values.astype(float) 
    
    # add dots variables
    # add categorical variables for skincolor, gender
    
    
    # ols approach
    try: 
        
        # cartesian coordinates
        y = prefs[np.newaxis].T
        X = sm.add_constant(np.vstack([affil, power, race, gender]).T)
        model = sm.OLS(y, X).fit()
        ols_df.loc[s,['socialValue_cartAffil_beta', 'socialValue_cartPower_beta',  
                      'socialValue_cartRace_beta', 'socialValue_cartGender_beta',
                      'socialValue_cart_bic']] = model.params[1:].tolist() + [model.bic]
        
        # polar coordinates (w/ self reference point)
        # - next: interaction?
        X = sm.add_constant(np.vstack([dist, cosine, race, gender]).T)
        model = sm.OLS(y, X).fit()
        ols_df.loc[s,['socialValue_polarDist_beta', 'socialValue_polarCos_beta', 
                      'socialValue_polarRace_beta', 'socialValue_polarGender_beta',
                      'socialValue_polar_bic']] = model.params[1:].tolist() + [model.bic]

        # polar coordinates (w/ self reference point)
        X = sm.add_constant(np.vstack([dist, race, gender]).T)
        model = sm.OLS(y, X).fit()
        ols_df.loc[s,['socialValue_dist_beta', 'socialValue_distRace_beta', 'socialValue_distGender_beta',
                      'socialValue_dist_bic']] = model.params[1:].tolist() + [model.bic]
        
    except:
        
        ols_df.loc[s,['socialValue_cartAffil_beta', 'socialValue_cartPower_beta',  
                      'socialValue_cartRace_beta', 'socialValue_cartGender_beta',
                      'socialValue_cart_bic']] = [np.nan, np.nan, np.nan, np.nan, np.nan]
        ols_df.loc[s,['socialValue_polarDist_beta', 'socialValue_polarCos_beta', 
                      'socialValue_polarRace_beta', 'socialValue_polarGender_beta',
                      'socialValue_polar_bic']] = [np.nan, np.nan, np.nan, np.nan, np.nan]
        ols_df.loc[s,['socialValue_dist_beta', 'socialValue_distRace_beta', 'socialValue_distGender_beta',
                      'socialValue_dist_bic']] = [np.nan, np.nan, np.nan, np.nan]

if 'socialValue_cartAffil_beta' in data.columns:
    data.update(ols_df)
else:
    data = data.merge(ols_df)      
```

## TODO: Can we predict the choice? Not v. well so far...
 - this would have to be at like 60+% at least to be interesting...

```python
preds = ['ego_dists_end_', 'affil_tendency_','power_tendency_', 'memory_']
X = []
y = []

# iterate over each trial
for t in np.arange(1, 16):
    
    trial_cols = [c for c in repl_df.columns if ('forced' in c) & ('_' + str(t) + '_' in c)]
    trial_chars = [c.split('_')[-1] for c in trial_cols][0:2]
    choice_sides = ((repl_df[[c for c in trial_cols if 'side' in c]] == 'R') * 1).values # L = 0, R = 1

    # iterate over each subject
    for s, (choices, side) in enumerate(zip(repl_df[trial_cols[0:2]].values, choice_sides)):
        try: # bc theres nans occasionally

            # which side/character was chosen
            y.append(side[0])
            chosen_ix = np.where(choices==np.max(choices))[0]
            chosen_char = trial_chars[int(chosen_ix)]
            unchosen_char = trial_chars[int(abs(int(chosen_ix) - 1))]

            # organize predictors 
            sub_df = repl_df.iloc[s, :]
            chosen_x = sub_df[[p + chosen_char for p in preds]].values.astype(int)
            unchosen_x = sub_df[[p + unchosen_char for p in preds]].values.astype(int)
            # have to infer which character was on which side... want the X to be L-R
            if side == 0: # chosen character is on left: chosen - unchosen 
                x = chosen_x - unchosen_x
            else: # chosen character is on right: unchosen - chosen 
                x = unchosen_x - chosen_x
            X.append(list(x))

        except:
            continue
```

```python
X = np.array(X)
y = np.array(y)
accs = []
for train_ix, test_ix in KFold(n_splits=5).split(X):    
    X_train = scipy.stats.zscore(X[train_ix])
    X_test = scipy.stats.zscore(X[test_ix])
    clf = LogisticRegression(random_state=0, solver='saga', penalty='l2').fit(X_train, y[train_ix])
    accs.append(clf.score(X_test, y[test_ix]))
print(f'acc={np.round(np.mean(accs), 2)}, chance={np.round(np.mean(y), 2)}')
```
