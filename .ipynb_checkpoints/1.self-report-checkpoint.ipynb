```python
import os, sys, glob, random, math, patsy 
import pandas as pd
import numpy as np
from sklearn import metrics
from scipy.spatial import distance
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

import warnings
warnings.filterwarnings("ignore")

#---------------------------------------------------------------------------

user = os.path.expanduser('~')
sys.path.insert(0, f'{user}/Dropbox/Projects/toolbox/toolbox')
from utils import merge_dfs
from dim_reduction import DimReduction


# add this to the utils file
character_roles = ['first', 'second', 'assistant', 'powerful', 'boss'] # order of matlab outputs...
demo_controls   = ['demo_age', 'iq_score', 'demo_gender_1W', 'demo_race_white']
ques_prefixes   = ['oci', 'zbpd', 'sds', 'aes', 'sss', 'lsas_av', 'apdis', 'bapq']
ques_labels     = ['OCD', 'Borderline', 'Depression', 'Apathy', 'Schizotypy', 'Social Anxiety', 'Social Avoidance', 'Autism'] # match prefixes


#---------------------------------------------------------------------------

# convenience functions
def subset_df(df, ques_prefixes):
    ques_dfs = []
    for ques in ques_prefixes: 
        ques_dfs.append(df.filter(regex=(f"{ques}_.*")))
    ques_df = pd.concat(ques_dfs, axis=1)
    ques_cols = [c for c in ques_df if ('_att' not in c) & ('score' not in c)]
    ques_df = ques_df[ques_cols]
    ques_df.insert(0, 'sub_id', df['sub_id'])
    return ques_df

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
main_dir = f'{user}/Desktop/SNT_data/SNT-online_behavioral/Data'

#------------------------------------------------------------
# initial sample
init_dir = f'{main_dir}/Initial_2021/Summary/'
init_df  = pd.read_excel(f'{init_dir}/All-data_summary_n613_clean.xlsx')

# corr mat for factor analyis [R's hetcor library]
init_corrmat = pd.read_csv(f'{init_dir}/Individual_summaries/Questionnaire_hetcor_n636.csv')
init_corrmat.set_index('Unnamed: 0', inplace=True)

print(f'Initial n={init_df.shape[0]}')

#------------------------------------------------------------
# replication sample
repl_dir = f'{main_dir}/Replication_2022/Summary/'
repl_df  = pd.read_excel(f'{repl_dir}/All-data_summary_n264_clean.xlsx')

# corr mat for factor analyis [R's hetcor library]
repl_corrmat = pd.read_csv(f'{repl_dir}/Individual_summaries/Questionnaire_hetcor_n276.csv')
repl_corrmat.set_index('Unnamed: 0', inplace=True)

print(f'Replication n={repl_df.shape[0]}')
```

# Exploratory Factor Analysis

## 1 - Run PCA to find optimal number of factors [initial sample only]

```python
init_ques_df = subset_df(init_df, ques_prefixes)

# pca
pca = DimReduction('pca')
pca.fit_transform(init_ques_df, num_comps=len(init_ques_df.columns)-1)

# scree plot for factor selection
fig, ax = plt.subplots(figsize=(5,5))
barlist = plt.bar(np.arange(0,15), pca.eigvals[0:15], width=1.0, facecolor='blue', edgecolor='black')
barlist[0].set_facecolor('orange')
barlist[1].set_facecolor('orange')
barlist[2].set_facecolor('orange')
plt.ylabel('Eigenvalue', fontsize=20)
plt.xlabel('Factor number', fontsize=20)
plt.xticks(np.arange(0,15), ['','','3','','','6','','','9','','','12','','','15'])
plt.show()

# eigenvalues is variance explained by each component
print('top 5 eigenvalues:')
print(pca.eigvals[0:5])
```

## 2 - Run factor analysis
#### heterogeneous corr matrix computed w/ hetcor in R

Resources: 
- https://www.ibm.com/docs/en/spss-statistics/28.0.0?topic=analysis-factor-rotation
- https://www.datacamp.com/community/tutorials/introduction-factor-analysis

```python
n_comps  = 3 
loading_thresh = 50 # define a threshold
rotations = ['quartimax', 'equamax', 'promax', 'oblimin', 'quartimin']
```

```python

for rotation in ['equamax']: 

    #-----------------------------------------------------------------
    # get questionnaire data and appropriate correlation matrix
    #-----------------------------------------------------------------


    init_ques_df = subset_df(init_df, ques_prefixes)
    items_mask   = init_corrmat.columns.isin(init_ques_df.columns)
    init_corrmat = init_corrmat.loc[items_mask, items_mask]


    #-----------------------------------------------------------------
    # run factor analysis
    #-----------------------------------------------------------------


    init_efa = DimReduction('efa')
    init_efa.fit_transform(init_ques_df, corr_matrix=init_corrmat, num_comps=n_comps, rotation=rotation) 
    # cm = efa_full.plot_corrmat()
    init_efa.plot_loadings()
    # init_efa.plot_components()

    # output the factor scores [!! make sure in correct order !!]
    # init_df[['factor_social', 'factor_mood', 'factor_compulsive']] = init_efa.X_reduced
    # weights = pd.DataFrame(repl_efa.loadings, columns=['factor01', 'factor02', 'factor03'])
    # weights.insert(0, 'question', repl_ques_df.columns[1:])


    #-----------------------------------------------------------------
    # run reduced factor analysis
    #-----------------------------------------------------------------


    # get the best items (largest loadings) from full efa
    min_items_ = init_efa.efa_partial(thresh = loading_thresh)
    init_min_items  = [i for l in min_items_[::-1] for i in l]
    items_mask = init_corrmat.columns.isin(init_min_items) 

    # grab the items for corrmat again
    init_corrmat_min = init_corrmat.loc[items_mask, items_mask]
    init_ques_df_min = init_ques_df.iloc[:, np.isin(init_ques_df.columns, init_min_items)][init_min_items]

    # run the factor analysis w/ best items
    init_efa_min = DimReduction('efa')
    init_efa_min.fit_transform(init_ques_df_min, corr_matrix=init_corrmat_min, num_comps=n_comps, rotation=rotation)
    # cm = efa_partial.plot_corrmat()
    # labels = ['OCD', 'Depression', 'Apathy', 'Schizotypy', 'Social Anxiety', 'Social Avoidance', 'Autism']
    # colors = get_unique(np.array(efa_full.item_colors)[items_mask])
    init_efa_min.plot_loadings()

    # output the factor scores [!! make sure in correct order !!]
    init_df[['factor_social_mood', 'factor_autism', 'factor_intrusive']] = init_efa_min.X_reduced
```
