```python
def random_colors(num_colors=10):
    from random import randint
    colors = ['#%06X' % randint(0, 0xFFFFFF) for i in range(num_colors)]
    return colors

def plot_ols_models_metrics(results_df, metric='train_BIC', colors=[]):
    
    # assign colors
    models = np.unique(results_df['model'])
    n_models = len(models)
    n_folds = list(results_df['model']).count(models[0])
    results_df['color'] = np.repeat(np.arange(n_models), n_folds, 0)
    
    if len(colors) == 0:
        colors = random_colors(num_colors=n_models)
        # color = iter(plt.cm.cool(np.linspace(0, 1, n_models))) # need more different colors
        # colors = []
        # for i in range(n_models): colors.append(list(next(color)))
    
    metric_df = results_df[results_df['metric'] == metric]
    mean_df = metric_df[metric_df['fold'] == 'mean']
    
    if 'IC' in metric: 
        ascending = True
    else:
        ascending = False 
    mean_df.sort_values('value', ascending=ascending, inplace=True)
    ordered_models = mean_df['model'].values
    ordered_colors = mean_df['color'].values

    fig, ax = plt.subplots(figsize=(15,7))
    ax = sns.barplot(x='model', y='value', hue='fold', data=metric_df, order=ordered_models)
    for i, bar in enumerate(ax.patches):
        c = i % n_models
        bar.set_color(colors[ordered_colors[c]])
        bar.set_edgecolor(".2")
    ax.set_ylabel(metric, fontsize=20)
    ax.set_ylim(np.min(metric_df['value'])-np.min(metric_df['value'])*.05, 
                np.max(metric_df['value'])+np.max(metric_df['value'])*.05)
    ax.set_xticklabels([x for x in ordered_models], rotation=60, ha='right')
    plt.legend([],[], frameon=False)
    plt.show()

def factor_model_metrics(models, df, factor_label, covariates, n_splits):
    res = []
    for model in models:
        if n_splits is None:
            ols_df, _ = run_ols(model, factor_label, df, covariates=covariates, n_splits=n_splits, plot=False)
        else:
            ols_df = run_ols(model, factor_label, df, covariates=covariates, n_splits=n_splits, plot=False)
        df_ = pd.melt(ols_df.reset_index(), id_vars='index', value_vars=['train-BIC', 'train-r_squared_adj', 'test-BIC', 'test-r_squared_adj', 'test-RMSE'], 
                     var_name='metric', value_name='value')
        df_.rename(columns={'index':'fold'}, inplace=True)
        df_.insert(0, 'model', (' + ').join(model))
        res.append(df_)
    res = pd.concat(res,0)
    return res
```

# Test main predictions: social specificity + contribution of both dimensions

Regress the behavior onto the factor scores 
- factor scores: each individual's score on each factor 
    - the linear combination of factor loadings and items 
    - factor loadings are the weight defining relationship between item & factor 


```python
#------------------------------------------------------------------------------------------------------------
# run OLS
#------------------------------------------------------------------------------------------------------------

# factors = ['factor_social_mood', 'factor_autism', 'factor_intrusive']
factors = ['factor_social_min','factor_mood_min','factor_compulsive_min']

samples = ['Initial', 'Replication']
demo_controls = ['demo_age', 'demo_gender_1W', 'demo_race_white', 'iq_score', 'demo_ladder_rate', 'reaction_time_mean'] # , 'demo_edu_level', 'demo_diag_6mo', 
models = [['affil_mean_mean_z', 'power_mean_mean_z']] # , ['pov_dist_mean_z']

res = []
for s, sample in enumerate(samples):
    df = data[data['sample']==s]
    for f, fname in enumerate(factors):
        for preds in models:

            _, ols_obj = run_ols(preds, f'{fname}', df, covariates=demo_controls, n_splits=None, plot=False)
            
            for p,pred in enumerate(preds): 
                
                if '*' in pred:
                    ci = ols_obj.conf_int().loc[pred.split(' * ')[0] + ':' + pred.split(' * ')[1]].values
                else: 
                    ci = ols_obj.conf_int().loc[pred].values
                    
                rsq = ols_obj.rsquared_adj
                beta = ols_obj.params[1+p]
                res.append([(' + ').join(preds), pred, sample, fname, rsq, beta, ci[0], ci[1], ols_obj.pvalues[1+p]])
                
ols_df = pd.DataFrame(res, columns=['model', 'predictor', 'sample', 'factor', 'r2_adj', 'beta', '95%_lb', '95%_ub', 'p-value'])
# display(ols_df)

#------------------------------------------------------------------------------------------------------------
# plot betas, cis & significance
#------------------------------------------------------------------------------------------------------------

plot_dict = {'affil_mean_mean_z':['Affiliation', "#4374B3"],
            'power_mean_mean_z':['Power', "#FF0B04"],
            'positive_tendency_mean_z':['Positive', '#3CA926'],
            'affil_mean_mean_z * power_mean_mean_z':['Affil*Power', '#6a329f'],
            'pov_dist_mean_z':['Distance', '#1ACBC0']}
            #  'ego2d_angles_all_mean_cos':['Angle', '#C823C6'], # hard to compare b/c on diff scale... maybe z-score instead?

task_preds = np.array(models).flatten().tolist()

# get colors
legend_labels = []
plot_colors   = []
for pred in task_preds:
    legend_labels.append(plot_dict[pred][0])
    plot_colors.append(plot_dict[pred][1])
plot_df = ols_df[ols_df['predictor'].isin(task_preds)]

pal = sns.set_palette(sns.color_palette(plot_colors))

with sns.axes_style("white"):
    
    # barplot
    c = sns.catplot(x="sample", y="beta", hue="predictor", col="factor", data=plot_df,
                    ci=None, legend=False, color=pal,
                    hue_order=task_preds, row_order=plot_colors, col_order=factors,
                    kind="bar", errcolor=".1", errwidth="1", height=15, aspect=.3,
                    edgecolor='black')    
    (c.set_ylabels('Beta (z)', fontsize=15)
      .set_xlabels('', fontsize=15)
      .set_xticklabels(['',''], fontsize=15) 
      .set(ylim=(-.36, .5))
      .set_titles()) 
    c.fig.set_figwidth(7)
    c.fig.set_figheight(7.5)

    ## plot the CIs & significance
    task_preds_ = np.repeat(task_preds, 2)
    for a, ax in enumerate(c.axes[0,:]): # diff factors
        for b, bar in enumerate(ax.patches):
            
            # replication bar positions
            if b % 2 == 1: bar.set_hatch('//')
                
            # get results
            ols_ = ols_df[(ols_df['sample'] == samples[b % 2]) 
                          & (ols_df['predictor'] == task_preds_[b]) 
                          & (ols_df['factor'] == factors[a])]
            beta = ols_['beta'].values
            ci   = np.abs((ols_['95%_ub'] - ols_['95%_lb']))/2
            pval = ols_['p-value'].values

            # add the CI
            bar_x_coords = bar.xy[0] + bar.get_width()/2  # where on x-axis to place ci bar
            ax.errorbar(bar_x_coords, beta, yerr=ci, ecolor='black', fmt=' ', zorder=2, elinewidth=1.5, dash_capstyle='butt') 
            
            # add the significance stars
            d = .012
            if pval < 0.05:
                ax.text(bar_x_coords, .29, '*', ha='center', va='bottom', color='black', size=30)  
            if pval < 0.01:
                ax.text(bar_x_coords, .29+d, '*', ha='center', va='bottom', color='black', size=30)   
            if pval < 0.005:
                ax.text(bar_x_coords, .29+2*d, '*', ha='center', va='bottom', color='black', size=30)   

    # legend
    handles, labels = plt.gca().get_legend_handles_labels()
    patch = mpatches.Patch(facecolor='white', edgecolor='black', hatch='//', label='Replication')
    handles.extend([patch])
    plt.legend(title='', loc='upper right', handles=handles, labels=legend_labels + ['Replication'],
               title_fontsize=13, fontsize=15, frameon=False, bbox_to_anchor=(2.25, 1), borderaxespad=0)
    
    
    
#---------------------------------------------------------------
# plot correlaitons
#---------------------------------------------------------------

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)

sns.regplot(x=data['affil_mean_mean'], y=data[factors[0]], 
            color="#4374B3", 
            scatter_kws={'alpha':0.33},
            ax=axs[0])
axs[0].set_xlabel('Affiliating', fontsize=15)
axs[0].set_ylabel('Factor', fontsize=15)
axs[0].set_xlim(-1,1)
axs[0].set_xticks([-1, -0.5, 0, 0.5, 1])

sns.regplot(x=data['power_mean_mean'], y=data[factors[0]], 
            color="#FF0B04", 
            scatter_kws={'alpha':0.33},
            ax=axs[1])
axs[1].set_xlabel('Giving power', fontsize=15)
axs[1].set_ylabel('', fontsize=15)
axs[1].set_xlim(-1,1)
axs[1].set_xticks([-1, -0.5, 0, 0.5, 1])

sns.regplot(x=data['pov_dist_mean'], y=data[factors[0]], 
            color="#1ACBC0",
            scatter_kws={'alpha':0.33},
            ax=axs[2])
axs[2].set_xlabel('Distance', fontsize=15)
axs[2].set_ylabel('', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
# ensure the mapping is correct: 
# x = np.zeros((13, 13))
# y = np.zeros((13, 13))
# for a_adj in np.arange(-6, 7):
#     for p_adj in np.arange(-6, 7): 
#         x[:, a_adj + 6] = a_adj # 2nd ix is cols/x 
#         y[6 - p_adj, :] = p_adj # 1st ix is rows/y


from mpl_toolkits.axes_grid1 import make_axes_locatable

# define the space
units = 13 # make odd (13 is regular)
space_corrs = np.zeros((units, units))
adj = int(units/2 - .5) # readjust back to plot
lb = -adj
ub = adj + 1

# split samples out
for s in range(3,4):
    
    if s == 3: df = data
    else:      df = data[data['sample'] == s]
    
    # get a new reference point
    vec_refs = pd.DataFrame()
    for a_adj in np.arange(lb, ub):
        for p_adj in np.arange(lb, ub):

            vecs = []
            for char in character_roles[0:5]:
                ap = np.concatenate((df['affil_coord_' + char].values.reshape(-1,1), 
                                     df['power_coord_' + char].values.reshape(-1,1)),1)
                ap_ref = ap - (a_adj, p_adj)
                vecs.append([np.linalg.norm(ap) for ap in ap_ref])

            # calculate correlation
            mean_vecs = np.mean(vecs, 0)
            vec_refs['ref_' + str(a_adj) + str(p_adj)] = mean_vecs
            r, p = scipy.stats.pearsonr(mean_vecs, df[factor])

            # put coefficient into ref point
            space_corrs[adj - p_adj, a_adj + adj] = r

    # visualize
    # space.astype(str) # 0 out uninteresting cells to control annotation (https://stackoverflow.com/questions/60611055/add-annotation-to-specific-cells-in-heatmap)
    fig, ax = plt.subplots(figsize=(6,6))
    im = plt.imshow(space_corrs, cmap='RdBu_r', zorder=0)
    # sns.heatmap(space_corrs, square=True, annot=False, cmap='RdBu_r', annot_kws={"size": 5}, linewidths=.1, zorder=0)
    ax.set_xticks(np.arange(0, units, 1))
    ax.set_yticks(np.arange(0, units, 1))
    ax.set_yticklabels(reversed(np.arange(lb, ub)))
    ax.set_xticklabels(np.arange(lb, ub))
    ax.set_xlabel('Affiliation', fontsize=20)
    ax.set_ylabel('Power', fontsize=20)
    
    # adding a colorbar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax, orientation='vertical')    
    cbar.ax.tick_params(labelsize=8) 
    #cbar.ax.set_ylabel('Correlation', rotation=270, fontsize=15)
    
    plt.tight_layout()
    plt.show()
    
    #     # add first eigenvector
    #     eig_vals, eig_vecs = np.linalg.eigh(space_corrs) # for symmetric matrices
    #     sort = np.argsort(eig_vals)[::-1]
    #     eig_vals = eig_vals[sort]
    #     eig_vecs = eig_vecs[:, sort]
    #     origin = [0,0]
    #     ax.quiver(*origin, *eig_vecs[0], color=['r'], scale=eig_vals[0], zorder=1)
    #     plt.show()

    # best reference frame coordinates
    # - the point at which if distance is measured from, it maximizes the correlation 
    best = np.where(space_corrs==np.max(space_corrs)) # best positive correlation
    best_r = -(best[0][0] - adj) # will be power (y)
    best_c = best[1][0] - adj # will be affil (x)
    print(best_c, best_r)

    # 
```

### Multivariate methods: canonical correlation analysis (CCA), canonical partial least squares (PLS) 
Can maybe use to compare our 2D spaces: compare two sets of variables, reduce dimensionality
- better to have a single model than a bunch of different models that are hard to compare against each other

CCA: 
- Goal is to find weights to max corr(X,Y), with X ~ a1 * x1 + a2 * x2 & Y ~ b1 * y1 + b2 * y2 

PLS: https://towardsdatascience.com/partial-least-squares-f4e6714452a
- good with multi-collinearity in the predictors (unlike multivariate regression)
- PLS disciminant analysis (PLS-DA) is when DVs are categorical
- PLS1 v PLS2: 2 means 2 dependent variables
- Can use a kernel trick to apply to non-linear problems: fit in a high-d space so input can be considered linear

CCA v PLS (v PCR)
- **PLS can capture non-linearities**: CCA maxes linear correlation, PLS maxes covariance which can capture non-linearities (& principle components regression focuses on variance)
- **PLS can estimate more components than dimensions in the data**:  CCA can only capture simpler relationships, PLS can capture more complex ones (arbitrary # of dims)
- **PLS is computationally simpler**: CCA uses eigendecomposition on correlation matrix; PLS uses regression
- **PLS v PCR (sklearn docs)**: PCR: PCA on training data + regressor on the transformed samples; "In PCA, the transformation is purely unsupervised, meaning that no information about the targets is used. As a result, PCR may perform poorly in some datasets where the target is strongly correlated with directions that have low variance. Indeed, the dimensionality reduction of PCA projects the data into a lower dimensional space where the variance of the projected data is greedily maximized along each axis. Despite them having the most predictive power on the target, the directions with a lower variance will be dropped, and the final regressor will not be able to leverage them.... PLS is both a transformer and a regressor... ***The main difference with PCR is that the PLS transformation is supervised." Therefore...it will use low variance directions if they are predictive of the target***

### PLS regression

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline

def plot_pca_2d(X):
    pca = PCA(n_components=2).fit(X)
    plt.scatter(X[:, 0], X[:, 1], alpha=0.3, label="samples")
    for i, (comp, var) in enumerate(zip(pca.components_, pca.explained_variance_)):
        comp = comp * var  # scale component by its variance explanation power
        plt.plot(
            [0, comp[0]],
            [0, comp[1]],
            label=f"Component {i}",
            linewidth=5,
            color=f"C{i + 4}")
    plt.gca().set(
        aspect="equal",
        title="Principal components",
        xlabel="first feature",
        ylabel="second feature",
    )
    plt.legend()
    plt.show()
    
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict


def optimise_pls_cv(X, y, n_comp, plot_components=True):
 
    '''Run PLS including a variable number of components, up to n_comp,
       and calculate MSE 
       Source: https://nirpyresearch.com/partial-least-squares-regression-python/
    '''
 
    
    n_targets = y.shape[1]
    
    # Cross-validated pls
    mse = []
    component = np.arange(1, n_comp)
    for i in component:
        pls = PLSRegression(n_components=i)
        y_cv = cross_val_predict(pls, X, y, cv=5) 
        mse.append(mean_squared_error(y, y_cv))
        # comp = 100*(i+1)/40
    
    #--------------------------------------------------
    # find minimum in MSE to find num of components
    #--------------------------------------------------
    
    msemin = np.argmin(mse)
    n_comps = msemin + 1   # optimal number of components 
    if plot_components is True:
        with plt.style.context(('ggplot')):
            fig = plt.figure(figsize=(3,3))
            plt.plot(component, np.array(mse), '-v', color = 'blue', mfc='blue')
            plt.plot(component[msemin], np.array(mse)[msemin], 'P', ms=10, mfc='red')
            plt.xlabel('Number of components')
            plt.ylabel('MSE')
            plt.title('PLS')
            plt.xlim(left=0)
            plt.show()
 
    # Calibration: entire dataset
    pls_cal = PLSRegression(n_components=n_comps)
    pls_cal.fit(X, y)
    X_tr_cal = pls_cal.transform(X) # transform X variables w/ the estimated weights that maximize covariance between X & Y
    y_preds_cal = pls_cal.predict(X)
 
    # Cross-validation
    pls_cv = PLSRegression(n_components=n_comps)
    y_preds_cv = cross_val_predict(pls_cv, X, y, cv=10)    
    
    # metrics
    score_c  = r2_score(y, y_preds_cal)
    score_cv = r2_score(y, y_preds_cv)
    mse_c  = mean_squared_error(y, y_preds_cal)
    mse_cv = mean_squared_error(y, y_preds_cv)
    
    print(f"Suggested number of components = {n_comps}")
    print(f'Calibration: R2 = {np.round(score_c,3)}, MSE = {np.round(mse_c,3)}')
    print(f'Cross-validation: R2 = {np.round(score_cv,3)}, MSE ={ np.round(mse_cv,3)}')

#     # Plot regression and figures of merit
#     rangey = max(y) - min(y)
#     rangex = max(y_c) - min(y_c) 
#     # Fit a line to the CV vs response
#     z = np.polyfit(y, y_c, 1)
#     with plt.style.context(('ggplot')):
#         fig, ax = plt.subplots(figsize=(9, 5))
#         ax.scatter(y_c, y, c='red', edgecolors='k')
#         #Plot the best fit line
#         ax.plot(np.polyval(z,y), y, c='blue', linewidth=1)
#         #Plot the ideal 1:1 line
#         ax.plot(y, y, color='green', linewidth=1)
#         plt.title('$R^{2}$ (CV): '+str(score_cv))
#         plt.xlabel('Predicted $^{\circ}$Brix')
#         plt.ylabel('Measured $^{\circ}$Brix')
#         plt.show()

    #--------------------------------------------------
    #--------------------------------------------------
    
    fig, axs = plt.subplots(n_comps, n_targets, figsize=(3*n_comps, 3*n_targets))
    for r in range(n_targets):
        for c in range(n_comps):
            axs[c,r].scatter(X_tr_cal[:,c], y[:,r], alpha=0.3, label="ground truth")
            axs[c,r].scatter(X_tr_cal[:,c], y_preds_cal[:,r], alpha=0.3, label="predictions")
            axs[c,r].set(xlabel=f"Projected X {c+1}", ylabel=f"True y {r+1}", title=f"Component {c+1}")
            axs[c,r].legend()
    plt.tight_layout()
    plt.show()

    return pls_cal
```

```python
yvars = ['affil_mean_mean', 'power_mean_mean']
xvars = ['sds_score', 'aes_score', 'oci_score', 'bapq_score', 'sss_score', 'lsas_av_score', 'apdis_score', 'zbpd_score']
nav  = data[yvars]
ques = data[xvars]
mask = np.sum(np.isfinite(ques), axis=1) == len(xvars)
nav  = scipy.stats.zscore(nav[mask])
ques = scipy.stats.zscore(ques[mask])

# X = ques.values
# plot_pca_2d(X)
pls_cal = optimise_pls_cv(ques.values, nav.values, len(xvars), plot_components=True)

#----------------------------------------------------------------------
# bi-plot
#----------------------------------------------------------------------

plt.figure(figsize=(5,5))
plt.xlim((-1.2,1.2))
plt.ylim((-1.2,1.2))
plt.title('Bi-plot', fontsize=18)

# plot the component scores: each subjects score on the components
#----------------------------------------------------------------------

# extract scores (one score per individual per component)
scores_df = pd.DataFrame(pls_cal.x_scores_)

# standardize scores between -1 and 1 so they fit on the plot
std_scores_dim1 = 2 * ( (scores_df[0] - min(scores_df[0])) / (max(scores_df[0]) - min(scores_df[0])) ) -1
std_scores_dim2 = 2 * ( (scores_df[1] - min(scores_df[1])) / (max(scores_df[1]) - min(scores_df[1])) ) -1
# scores = np.vstack([std_scores_dim1.values, std_scores_dim1.values]).T

plt.scatter(std_scores_dim1.values, std_scores_dim2.values, color='grey')

# if want arrows instead
# for i in range(len(std_scores_dim1)):
#     x = std_scores_dim1[i]
#     y = std_scores_dim2[i]
#     plt.arrow(0, 0, x, y, color='grey')
    # plt.text(x, y, data.loc[i, 'countries'], color= 'blue')


# loadings vector plot: direction vectors that define the pls model
# - how do the original variables contribute to the components
#----------------------------------------------------------------------

# extract loadings (one loading per variable per component)
loadings_df = pd.DataFrame(pls_cal.x_loadings_)
for i in range(loadings_df.shape[0]):
    x = loadings_df.loc[i, 0]
    y = loadings_df.loc[i, 1]
    plt.arrow(0, 0, x, y, color='red') 
    # plt.text(x, y, var_i, color='red', fontsize=15)

plt.xlabel('Component 1 score', fontsize=15)
plt.ylabel('Component 2 score', fontsize=15)
plt.show()
```
