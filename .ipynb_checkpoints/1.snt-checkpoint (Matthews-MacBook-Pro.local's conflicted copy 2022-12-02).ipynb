```python
# asymmetrical correlation matrices
def get_asymmetrical_corrmat(x_df_, y_df_, linear_correlation=scipy.stats.pearsonr, permutations=None):
    '''
        x_df & y_df should dfs of shape num_subs x num_variables
        robust to diff. amounts of missingness for diff. variables
        will do nonparametric correlations if specified & angular-angular & angular-linear correlations 
        also will do permutation testing
    '''
    
    x_df = x_df_.copy()
    y_df = y_df_.copy()

    y_cols = y_df.columns
    x_cols = x_df.columns
    
    # avoid problems with data type
    x_df[x_cols] = x_df[x_cols].apply(pd.to_numeric, downcast='float', errors='coerce')
    y_df[y_cols] = y_df[y_cols].apply(pd.to_numeric, downcast='float', errors='coerce')
    
    # turn any angles into radians...
    y_angles = ['angles' in c for c in y_cols]
    x_angles = ['angles' in c for c in x_cols]
    y_df[y_cols[y_angles]] = y_df[y_cols[y_angles]].apply(np.deg2rad) 
    x_df[x_cols[x_angles]] = x_df[x_cols[x_angles]].apply(np.deg2rad) 

    # for outputting
    coefs = np.zeros((len(y_cols), len(x_cols)))
    pvals = np.zeros((len(y_cols), len(x_cols)))
    
    for yi, y_col in enumerate(y_cols):
        for xi, x_col in enumerate(x_cols):

            # screen out subs w/ nans & infs 
            finite_mask = np.isfinite(x_df[x_col].values) * np.isfinite(y_df[y_col].values) 
            x = x_df[x_col][finite_mask]
            y = y_df[y_col][finite_mask]
            
            # correlation 
            if ('angles' not in x_col) & ('angles' not in y_col):
                correlation = linear_correlation 
            elif (('angles' in x_col) & ('angles' not in y_col)):
                correlation = circ_corrcl
            elif (('angles' not in x_col) & ('angles' in y_col)):
                correlation = circ_corrcl 
                y = x_df[x_col][finite_mask] # flip the order of circular & linear variables
                x = y_df[y_col][finite_mask]
            elif ('angles' in x_col) & ('angles' in y_col):
                correlation = circ_corrcc
            coefs[yi,xi], pvals[yi,xi] = correlation(x, y)
            
            # permutation significance testing
            if permutations: 
                perm_coefs = np.zeros(permutations)
                y_ = y.copy().values
                for n in range(permutations):
                    random.shuffle(y_)
                    perm_coefs[n], _ = correlation(x, y_)
                pvals[yi,xi] = get_perm_pval(coefs[yi,xi], perm_coefs, test='2-sided')
                
    corr_mat = pd.DataFrame(coefs, index=y_cols, columns=x_cols)
    pval_mat = pd.DataFrame(pvals, index=y_cols, columns=x_cols)
    return corr_mat, pval_mat


def plot_asymmetrical_corrmat(corr_mat, pval_mat, figsize=(15,15)):
    '''
    '''
    x_names  = corr_mat.columns
    y_names  = corr_mat.index
    corr_mat = corr_mat.values.copy()
    pval_mat = pval_mat.values.copy()

    # masked correlation plot
    masked_corrs = corr_mat * ((pval_mat < 0.05) * 1) 
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(masked_corrs, cmap="RdBu_r", vmax=1, vmin=-1) #np.max(masked_corrs)
    plt.grid(False)
    
    # 0.05: *, 0.01: **, 0.001: ***, <0.0001: ****
    for i in np.arange(len(y_names)):
        for j in np.arange(len(x_names)):
            if corr_mat[i,j] > 0.50:
                color = 'white'
            else:
                color = 'black'
            if pval_mat[i,j] < 0.0001:
                sig = '****'
            elif pval_mat[i,j] < 0.001:
                sig = '***'
            elif pval_mat[i,j] < 0.01:
                sig = '**'
            elif pval_mat[i,j] < 0.05:
                sig = '*'
            else:
                sig = ' '
            text = ax.text(j, i, sig, ha="center", va="center", color=color, fontsize=20)
            
    im_ratio = corr_mat.shape[0]/corr_mat.shape[1] # (height_of_image / width_of_image)
    # cbar = plt.colorbar(ax.get_children()[-2], ax=ax, orientation='vertical', fraction=0.044*im_ratio)
    # cbar.set_label(label='Correlation', size=20)
    # cbar.ax.tick_params(labelsize=9)

    ax.set_yticks(np.arange(len(y_names)))
    ax.set_yticklabels(y_names, fontsize=15)
    ax.set_xticks(np.arange(len(x_names)))
    ax.set_xticklabels(x_names, fontsize=15, rotation=90)

    # add grid lines
    ax.set_xticks(np.arange(-.5, len(x_names), 1), minor=True)
    ax.set_yticks(np.arange(-.5, len(y_names), 1), minor=True)
    ax.grid(which='minor', color='black', linestyle='-', linewidth=1.5)
    
    ax.set_title('Correlations', fontsize=20)

    return fig
def plot_space_density(x, y, figsize=(5,5), regression=False):
    
    '''
    '''
    
    xy = np.vstack([x, y])
    z  = gaussian_kde(xy)(xy) # point density

    # sort so that the bright spots are on top
    idx = z.argsort()
    x, y, z = x[idx], y[idx], z[idx]

    # plot
    fig, axs = plt.subplots(2, 2, figsize=figsize, 
                                  gridspec_kw={'hspace': 0, 'wspace': 0,
                                               'width_ratios': [5, 1], 'height_ratios': [1, 5]})
    axs[0,0].axis("off")
    axs[0,1].axis("off")
    axs[1,1].axis("off")
    sns.distplot(x, bins=20, ax=axs[0,0], color="Black")
    sns.distplot(y, bins=20, ax=axs[1,1], color="Black", vertical=True)
    axs[1,0].axhline(y=0, color='black', linestyle='-', zorder=-1)
    axs[1,0].axvline(x=0, color='black', linestyle='-', zorder=-1)
    axs[1,0].scatter(x, y, c=z, s=100)
    
    if regression: sns.regplot(x, y, scatter=False, color='black', ax=axs[1,0])
    
    axs[1,0].set_ylim([-1,1])
    axs[1,0].set_yticks([-1,-.5,0,.5,1])
    axs[1,0].set_xlim([-1,1])
    axs[1,0].set_xticks([-1,-.5,0,.5,1])
    
    plt.show()
```

## 2D placement (dots)

```python
dists_df = pd.DataFrame()
for s, data_ in data.iterrows(): 
    
    dists_df.loc[s, 'sub_id'] = data_['sub_id']
    
    # organize x,y - only the 5 real characters
    task_data = data_[['affil_mean_first','power_mean_first',
                       'affil_mean_second','power_mean_second',
                       'affil_mean_assistant','power_mean_assistant',
                       'affil_mean_powerful','power_mean_powerful',
                       'affil_mean_boss','power_mean_boss']].values
    dots_data = data_[['first_dots_affil','first_dots_power',
                       'second_dots_affil','second_dots_power',
                       'assistant_dots_affil','assistant_dots_power',
                       'powerful_dots_affil','powerful_dots_power',
                       'boss_dots_affil','boss_dots_power']].values
    
    
    #----------------------------------------------------------------------------------------
    # CHARACTER BY CHARACTER
    #----------------------------------------------------------------------------------------

    ori_xy  = np.array([[1,0]])
    task_xy = task_data.reshape(5,2).astype(float)
    dots_xy = dots_data.reshape(5,2).astype(float)    
    
    #--------------------------------------------
    # direct dots to task distances
    #--------------------------------------------

    
    # 1d distances
    dists_1d = task_xy - dots_xy 
    dists_df.loc[s, [f'taskVdots_affil_{role}' for role in character_roles[0:5]]] = dists_1d[:,0] # do dots get larger afiliation value?
    dists_df.loc[s, [f'taskVdots_power_{role}' for role in character_roles[0:5]]] = dists_1d[:,1] # no real hypothesis
    
    # 2d distances
    dists_df.loc[s, [f'taskVdots_dist_{role}' for role in character_roles[0:5]]] = pairwise_distances(task_xy, dots_xy).diagonal()

    
    #--------------------------------------------
    # diff in distances to common reference
    #--------------------------------------------
    # ori = np.array((6, -4)) / 6
    
    task_dist  = pairwise_distances(task_xy, ori_xy)
    dots_dist  = pairwise_distances(dots_xy, ori_xy)
    
    dists_df.loc[s, [f'dots_pov_dist_{role}' for role in character_roles[0:5]]]           = dots_dist.flatten()
    dists_df.loc[s, [f'taskVdots_pov_dist_diff_{role}' for role in character_roles[0:5]]] = task_dist.flatten() - dots_dist.flatten()

    # calculate_angle(task_pos - ori, dots_pos - ori,force_pairwise=False, direction=None)
    
    
    #----------------------------------------------------------------------------------------
    # WHOLE SHAPE
    #----------------------------------------------------------------------------------------

    
    #-----------------------------------------------------------
    # RSA: pairwise distances 
    #-----------------------------------------------------------
    
    # 1d
    affil_task_rdv = symm_mat_to_ut_vec(pairwise_distances(task_xy[:,0][:,np.newaxis]))
    affil_dots_rdv = symm_mat_to_ut_vec(pairwise_distances(dots_xy[:,0][:,np.newaxis]))
    dists_df.loc[s, f'taskVdots_affil_dist_tau'] = scipy.stats.kendalltau(affil_dots_rdv, affil_task_rdv)[0]

    power_task_rdv = symm_mat_to_ut_vec(pairwise_distances(task_xy[:,1][:,np.newaxis]))
    power_dots_rdv = symm_mat_to_ut_vec(pairwise_distances(dots_xy[:,1][:,np.newaxis]))
    dists_df.loc[s, f'taskVdots_power_dist_tau'] = scipy.stats.kendalltau(power_dots_rdv, power_task_rdv)[0]
        
    # 2d
    origins  = {'neu': [0,0], 'pov': [1,0]}
    metrics  = ['cosine', 'euclidean', 'cityblock']
    for ori, ori_xy in origins.items():
        for m, metric in enumerate(metrics): 
            # rsa: distance across characters
            dots_rdv = symm_mat_to_ut_vec(pairwise_distances(dots_xy - ori_xy, metric=metric))
            task_rdv = symm_mat_to_ut_vec(pairwise_distances(task_xy - ori_xy, metric=metric))
            dists_df.loc[s, f'taskVdots_{ori}_{metric}_dist_tau'] = scipy.stats.kendalltau(dots_rdv, task_rdv)[0]
            
    #-----------------------------------------------------------
    # areas etc
    #-----------------------------------------------------------
    
    # area = scipy.spatial.ConvexHull(dots_xy).volume
    
    
    #-----------------------------------------------------------
    # procrustes transformation
    # - transfroms one shape to fit the other
    # - the disparity is point-wise euclidean distance it is trying to minimize
    #-----------------------------------------------------------

    dists_df.loc[s, 'taskVdots_procDisp'] = scipy.spatial.procrustes(dots_xy, task_xy)[2] # disparity shouldnt matter on order

    #-----------------------------------------------------------
    # not sure what orthogonal procrustes is
    #-----------------------------------------------------------
    
    # _, scale = scipy.linalg.orthogonal_procrustes(task_xy, dots_xy) # (to_map, map_to)

    #-----------------------------------------------------------
    # mean elementwise euclidean distance from dots to task
    #-----------------------------------------------------------

    dists_df.loc[s, 'taskVdots_dist_mean'] = np.mean(scipy.spatial.distance.cdist(task_xy, dots_xy, 'euclidean').diagonal())

    #-----------------------------------------------------------
    # use least squares to get transform from
    # -- task @ transform = dots
    #-----------------------------------------------------------

    # - this returns a residual for both dims (i think).... could be useful?
    _, res, _ , _ = np.linalg.lstsq(task_xy, dots_xy, rcond=None)
    try: 
        dists_df.loc[s, 'taskVdots_lstsqRes_affil'] = res[0]
        dists_df.loc[s, 'taskVdots_lstsqRes_power'] = res[1]
    except:
        dists_df.loc[s, 'taskVdots_lstsqRes_affil'] = np.nan
        dists_df.loc[s, 'taskVdots_lstsqRes_power'] = np.nan
        
        
# means of elementwise comparisons
dists_df['dots_pov_dist_mean']      = np.mean(dists_df[[c for c in dists_df.columns if ('dots_pov_dist_' in c) & ('task' not in c)]], 1)
dists_df['taskVdots_pov_dist_mean'] = np.mean(dists_df[[c for c in dists_df.columns if 'taskVdots_pov_dist_' in c]], 1)
dists_df['taskVdots_dist_mean']     = np.mean(dists_df[[c for c in dists_df.columns if 'taskVdots_dist_' in c]], 1)
dists_df['taskVdots_power_mean']    = np.mean(dists_df[[c for c in dists_df.columns if 'taskVdots_power_' in c]], 1)
dists_df['taskVdots_affil_mean']    = np.mean(dists_df[[c for c in dists_df.columns if 'taskVdots_affil_' in c]], 1)
dists_df.to_excel(f'../../taskVdots_n{len(dists_df)}.xlsx', index=False)

# data = data.merge(dists_df)
# dists_df.head(3)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()
def word_count(string):
    return(len(string.split(" ")))

for role in character_roles[0:5]:
    free_resps = data[f'free_response_{role}'].values
    num_words, negative, positive, neutral = [], [], [], []
    for fr in free_resps:
        try:    
            num_words.append(word_count(fr))
            polarity = sentiment.polarity_scores(fr)
            negative.append(polarity['neg'])
            positive.append(polarity['pos'])
            neutral.append(polarity['neu'])
        except: 
            num_words.append(np.nan)
            positive.append(np.nan)
            negative.append(np.nan)
            neutral.append(np.nan)

    data[f'free_response_{role}_num_words'] = num_words
    data[f'free_response_{role}_positive'] = positive
    data[f'free_response_{role}_negative'] = negative
    data[f'free_response_{role}_neutral'] = neutral

data['free_response_num_words_mean'] = np.mean(data[[c for c in data.columns if ('free' in c) & ('num_words' in c)]], axis=1)
data['free_response_positive_mean'] = np.mean(data[[c for c in data.columns if ('free' in c) & ('positive' in c) & ('sia' not in c)]], axis=1)
data['free_response_negative_mean'] = np.mean(data[[c for c in data.columns if ('free' in c) & ('negative' in c) & ('sia' not in c)]], axis=1)
data['free_response_neutral_mean'] = np.mean(data[[c for c in data.columns if ('free' in c) & ('neutral' in c) & ('sia' not in c)]], axis=1)
```
