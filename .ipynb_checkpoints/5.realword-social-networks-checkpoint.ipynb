```python
def bootstrapped_ols(y, X, num_samples=1000, conf=95):
    '''
    '''
    
    sample_size = len(X)
    bb = []
    for b in range(num_samples):
        ixs_ = np.random.choice(range(sample_size), (sample_size)) # ixs for random samples 
        X_ = X.iloc[ixs_,:]
        y_ = y[ixs_]
        ols_obj = sm.OLS(y_, X_, missing='drop').fit()
        bb.append(ols_obj.params[1])
    bb.sort()
    lb = (100-conf)/2
    ub = 100-((100-conf)/2)
    ci = np.percentile(bb, [lb,ub]) 
    return {'betas': np.array(bb), 'ci': ci}
```

### behavior predicting network scores

```python
predictors_ = [['ego_dists_end_mean'], 
              ['affil_tendency_mean', 'power_tendency_mean'],
              ['affil_tendency_mean']]
sni_scores = ['sni_number_ppl'] # , 'sni_network_diversity']
bics = []
ols_objs = {}
for s, sni in enumerate(sni_scores): 
    bics_ = []
    for predictors in predictors_:
        for d, df in enumerate([orig_df, repl_df]):
            sni_mask = np.isfinite(df['sni_number_ppl'])
            X = sm.add_constant(df[predictors][sni_mask])
            X.reset_index(drop=True, inplace=True)
            y = df[sni][sni_mask]
            y.reset_index(drop=True, inplace=True)
            ols_obj = sm.OLS(y, X, missing='drop').fit()
            ols_objs[('+').join(predictors)] = ols_obj
            bics_.append(ols_obj.bic)
    bics.append(bics_)
```

### network scores predicting behavior

```python
outcomes = ['ego_dists_end_mean', 'affil_tendency_mean', 'power_tendency_mean']
sni_scores = ['sni_number_ppl', 'sni_network_diversity']
bic_df = pd.DataFrame(columns=['predicted', 'original', 'replication', 'mean'])
for o, outcome in enumerate(outcomes): 
    bics_ = []
    for d, df in enumerate([orig_df, repl_df]):
        sni_mask = np.isfinite(df['sni_number_ppl'])
        X = sm.add_constant(scipy.stats.zscore(df[sni_scores][sni_mask]))

        # X = sm.add_constant(pd.concat([scipy.stats.zscore(df[sni_scores][sni_mask]), df[demo_controls][sni_mask]], axis=1))
        X.reset_index(drop=True, inplace=True)
        y = scipy.stats.zscore(df[outcome][sni_mask])
        y.reset_index(drop=True, inplace=True)
        ols_obj = sm.OLS(y, X, missing='drop').fit()

        bics_.append(ols_obj.bic)
    bic_df.loc[o,:] = [outcome] + bics_ + [np.mean(bics_)]
bic_df
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax = sns.barplot(x='predicted', y='mean', data=bic_df, ci=None, palette='Blues_d', edgecolor=".4")
ax.set_ylabel('BIC', fontsize=20)
ax.set_ylim(np.min(bic_df['mean']) - np.min(bic_df['mean']) * .005, 
            np.max(bic_df['mean']) + np.max(bic_df['mean']) * .005)
ax.set_xticklabels(['Social distance', 'Affiliation', 'Power'], rotation=60, ha='right')
plt.show()
```
