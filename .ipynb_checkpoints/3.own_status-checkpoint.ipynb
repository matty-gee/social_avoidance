### Relationship betwn social judgments & both behavior & social avoidance improves if affiliation & power represented with diff relational structure 
- Affiliation: close to far - the average location between self & other
- Power: above or below - the difference in location between self & other

```python
data['likability_selfOther_avg']  = np.mean(data[['likability_mean', 'self_likability']], 1)
data['impact_selfOther_diff']     = data['impact_mean'].values - data['self_impact'].values
```

```python
affil_proxys  = ['likability_mean', 'self_likability', 'likability_selfOther_avg']
powers_proxys = ['impact_mean', 'self_impact', 'impact_selfOther_diff']

fontsize = 13
affil_color = "blue"
power_color = "red"

### affiliation ###
for yi, y in enumerate(['factor_social_min', 'affil_mean_mean']):
    fig, axs = plt.subplots(1, 3, figsize=(7.5, 2.5), sharey=True)
    for xi, x in enumerate(affil_proxys):
        sns.regplot(x=data[x], y=data[y], 
                    scatter_kws={'alpha':0.3}, color=affil_color,
                    ax=axs[xi])
        # axs[yi,0].set_ylim(-3,3)
        if xi == 0:
            axs[xi].set_ylabel(y, fontsize=fontsize)
        else:
            axs[xi].set_ylabel(' ', fontsize=fontsize)
        axs[xi].set_xlabel(x, fontsize=fontsize)
        add_sig_stars(x, y, data, demo_controls, ax=axs[xi], color=affil_color)
    fig.align_ylabels(axs)
    plt.tight_layout()
    plt.show()

### power ###
for yi, y in enumerate(['factor_social_min', 'power_mean_mean']):
    fig, axs = plt.subplots(1, 3, figsize=(7.5, 2.5), sharey=True)
    for xi, x in enumerate(powers_proxys):
        sns.regplot(x=data[x], y=data[y], 
                    scatter_kws={'alpha':0.3}, color=power_color,
                    ax=axs[xi])
        # axs[yi,1].set_ylim(-3,3)
        if xi == 0:
            axs[xi].set_ylabel(y, fontsize=fontsize)
        else:
            axs[xi].set_ylabel(' ', fontsize=fontsize)
        axs[xi].set_xlabel(x, fontsize=fontsize)
        add_sig_stars(x, y, data, demo_controls, ax=axs[xi], color=power_color)
    fig.align_ylabels(axs)
    plt.tight_layout()
    plt.show()

# ### distance x affil ###
# fig, axs = plt.subplots(1, 4, figsize=(10,2.5))
# color = "purple"
# for xi, x in enumerate(affil_proxys):
#     for yi, y in enumerate(['pov_dist_mean']):
#         sns.regplot(x=all_df[x], y=all_df[y], 
#                     scatter_kws={'alpha':0.3}, color=color,
#                     ax=axs[xi])
#         # axs[yi,1].set_ylim(-3,3)
#         if xi == 0:
#             axs[xi].set_ylabel(y, fontsize=fontsize)
#         else:
#             axs[xi].set_ylabel(' ', fontsize=fontsize)
#         axs[xi].set_xlabel(x, fontsize=fontsize)
#         add_sig_stars(x, y, all_df, demo_controls, ax=axs[xi], color=color)
# plt.tight_layout()
# plt.show()

# ### distance x power ###
# fig, axs = plt.subplots(1, 4, figsize=(10,2.5))
# color = "purple"
# for xi, x in enumerate(powers_proxys):
#     for yi, y in enumerate(['pov_dist_mean']):
#         sns.regplot(x=all_df[x], y=all_df[y], 
#                     scatter_kws={'alpha':0.3}, color=color,
#                     ax=axs[xi])
#         # axs[yi,1].set_ylim(-3,3)
#         if xi == 0:
#             axs[xi].set_ylabel(y, fontsize=fontsize)
#         else:
#             axs[xi].set_ylabel(' ', fontsize=fontsize)
#         axs[xi].set_xlabel(x, fontsize=fontsize)
#         add_sig_stars(x, y, all_df, demo_controls, ax=axs[xi], color=color)
# plt.tight_layout()
# plt.show()
```
